```python
# Q1. Analyze the distribution of psychological health indicators (e.g., stress, anxiety,
#     depression) across the population

fig, axes = plt.subplots(1, 3, figsize=(15,4))
sns.histplot(df["Stress_Score"], kde=True, ax=axes[0], color="tomato")
axes[0].set_title("Stress Score")

sns.histplot(df["Anxiety_Score"], kde=True, ax=axes[1], color="skyblue")
axes[1].set_title("Anxiety Score")

sns.histplot(df["Depression_Score"], kde=True, ax=axes[2], color="green")
axes[2].set_title("Depression Score")

plt.tight_layout()
plt.show()

print("Skewness:")
```

```python
# Q2. Explore the relationship between sleep duration and mental health scores.

sns.lmplot(x="Sleep_Duration", y="Stress_Score", data=df, scatter_kws={'alpha':0.4})
plt.title("Sleep Duration vs Stress Score")
plt.show()

sns.lmplot(x="Sleep_Duration", y="Anxiety_Score", data=df, scatter_kws={'alpha':0.4})
plt.title("Sleep Duration vs Anxiety Score")
plt.show()

sns.lmplot(x="Sleep_Duration", y="Depression_Score", data=df, scatter_kws={'alpha':0.4})
plt.title("Sleep Duration vs Depression Score")
plt.show()
```

```python
# Q3. Investigate the impact of physical activity on overall happiness and psychological well
#     being.
#     Physical activity vs Happiness

sns.scatterplot(x="Physical_Activity", y="Happiness_Index",
                hue="Stress_Score", data=df, palette="coolwarm")
plt.title("Physical Activity vs Happiness Index")
plt.show()
```

```python
# Q4. Compare age groups in terms of stress and depression levels.
#     Age groups vs Stress/Depression

bins   = [0,18,30,45,60,120]
labels = ["<18","18-30","30-45","45-60","60+"]
df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels)

sns.boxplot(x="Age_Group", y="Stress_Score", data=df)
plt.title("Stress Score by Age Group")
plt.show()

sns.boxplot(x="Age_Group", y="Depression_Score", data=df)
plt.title("Depression Score by Age Group")
plt.show()
```

```python
# Q5. Study the effect of work-related factors such as job stress and workload on anxiety or
#     depression levels.
#    Job stress/workload vs Anxiety/Depression

sns.scatterplot(x="Workload", y="Anxiety_Score",
                hue="Job_Stress_Level", data=df)
plt.title("Workload vs Anxiety Score")
plt.show()

sns.scatterplot(x="Workload", y="Depression_Score",
                hue="Job_Stress_Level", data=df)
plt.title("Workload vs Depression Score")
plt.show()
```
